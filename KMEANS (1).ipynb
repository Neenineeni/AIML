```python
#standardisation
x = df.iloc[:, 0:3].values
sc=MinMaxScaler()
x=sc.fit_transform(x)
x
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
from sklearn.cluster import KMeans
wcss=[] #within cluster sum of square
for i in range(1, 11):
    kmeans=KMeans(n_clusters=i, init='k-means++',max_iter=300,random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('no.of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
y_kmeans=kmeans.fit(x)
y_kmeans
```

```python
kmeans.cluster_centers_
```

```python
kmeans=KMeans(n_clusters=3)
#kmeans.fit(x)
y_kmeans=kmeans.fit_predict(x)
y_kmeans
```

```python
#visualize the clusters
plt.scatter(x[y_kmeans ==0, 0], x[y_kmeans == 0, 1], s = 100, c  = 'green',label='df.setosa')
plt.scatter(x[y_kmeans ==1, 0], x[y_kmeans == 1, 1], s = 100, c  = 'red',label='df.verisolor')
plt.scatter(x[y_kmeans ==2, 0], x[y_kmeans == 2, 1], s = 100, c  = 'blue',label='df.virginica')
plt.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:, 1], s=100, c = 'orange', label='Centroids') 
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.legend()
plt.show()
```
