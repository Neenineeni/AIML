```python
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
#data points 
data_points = np.array([
    [0.40, 0.53],
    [0.22, 0.38],
    [0.35, 0.32],
    [0.26, 0.41],
    [0.08, 0.41],
    [0.45, 0.30]
])
labels = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6']
```

```python
#PERFORM HEIRARCHICAL CLUSTERING
linked = sch.linkage(data_points, method='single', metric='euclidean')
```

```python
#plot dendogram
plt.figure(figsize=(10, 5))
sch.dendrogram(linked,labels=labels,orientation='top',distance_sort='ascending',show_leaf_counts=True)
plt.title("Agglomerative Clustering Dendrogram(single linkage)")
plt.xlabel("Sample Points")
plt.ylabel("Distance")
plt.grid(True)
plt.tight_layout()
plt.show()
```
